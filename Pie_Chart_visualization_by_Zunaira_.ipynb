```python
import matplotlib.pyplot as plt

# Questions and options for the quiz
questions = [
    {
        "question": "What is the capital of Pakistan?",
        "options": ["Muree", "Sukkur", "Karachi", "Islamabad"],
        "answer": "Islamabad"
    },
    {
        "question": "What is the currency of Pakistan?",
        "options": ["Euro", "Rupees", "Yen", "Pound"],
        "answer": "Rupees"
    },
    {
        "question": "In which of the following known as the brain of computer'?",
        "options": ["Mouse", "Keyboard", "C.P,U", "Data"],
        "answer": "C.P.U"
    },
    {
        "question": "What is the national flower of Pakistan?",
        "options": ["Rose", "Tuplip", "Daisy", "Jasmine"],
        "answer": "Jasmine"
    },
    {
        "question": "Which element has the chemical symbol 'O'?",
        "options": ["Oxygen", "Gold", "Osmium", "Ozone"],
        "answer": "Oxygen"
    }
]

# Variables to track score
correct_answers = 0
incorrect_answers = 0

# Function to ask questions
def ask_question(q):
    global correct_answers, incorrect_answers
    print(q['question'])
    for i, option in enumerate(q['options'], 1):
        print(f"{i}. {option}")

    try:
        # Input from user
        user_answer = int(input("Please enter the number of your answer: "))
        if user_answer < 1 or user_answer > len(q['options']):
            print("Invalid choice! Please choose a valid number.")
            return ask_question(q)  # Retry if invalid choice is made
    except ValueError:
        print("Invalid input! Please enter a number.")
        return ask_question(q)  # Retry if non-integer input is entered

    if q['options'][user_answer - 1] == q['answer']:
        print("Correct!\n")
        correct_answers += 1
    else:
        print(f"Incorrect! The correct answer was: {q['answer']}\n")
        incorrect_answers += 1

# Loop to ask each question
for question in questions:
    ask_question(question)

# Summary and visualization
print(f"\nYour score: {correct_answers} correct, {incorrect_answers} incorrect.")

# Visualization using matplotlib
labels = ['Correct', 'Incorrect']
sizes = [correct_answers, incorrect_answers]
colors = ['green', 'red']
explode = (0.1, 0)  # Explode the "Correct" slice

plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Quiz Results")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Save the pie chart as an image
plt.savefig("quiz_results.png")
plt.show()

# Save the results to a text file
with open("quiz_results.txt", "w") as file:
    file.write(f"Total Questions: {len(questions)}\n")
    file.write(f"Correct Answers: {correct_answers}\n")
    file.write(f"Incorrect Answers: {incorrect_answers}\n")

print("\nResults have been saved to 'quiz_results.txt' and the pie chart to 'quiz_results.png'.")
```
